## Function for Count Plot

```python
def plot_genral_count(column_name , title):
    '''
    inputs: column_name: to plot
            title: for the plot
    output: a count plot for the data in genral
            
    '''
    plot_genral_count.fig, plot_genral_count.ax = plt.subplots(figsize=(20 , 10))
    cmap = sns.cubehelix_palette(12  , reverse=True)
    sns.countplot(x=df[column_name], ax=plot_genral_count.ax , order=df[column_name].value_counts().index , palette=cmap)
    plt.xticks(rotation=15)
    plt.title(title)
```
